```python
label=['ADF Test Static','P value','Lag used','Number Of abs']
```

```python
#H0=not stationary
#H1=Stationary
def check(data):
    test=adfuller(data)
    print("5%= "+str(test[4]['5%']))
    for v,l in zip(test,label):
        print(l+" "+str(v))
        
    if(test[1]<=0.05):
        print("reject null, i.e it is stationary")
    else:
        print("accept null, i.e it not is stationary")
for i in df.columns:
    print("\n")
    print("For "+i)
    print("\n")
    print("\n")
    check(df[i])
```
